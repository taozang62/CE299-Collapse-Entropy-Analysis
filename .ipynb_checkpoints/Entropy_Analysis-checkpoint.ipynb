```python
%cd /Users/taozang/Zillow/Extracted_Data/12_ZAsmt_CSV/
import entropy_utility as ut
import pandas as pd
tables = ['Building', 'BuildingAreas', 
          'CareOfName', 'ExteriorWall', 'ExtraFeature', 'Garage', 'InteriorFlooring', 
          'InteriorWall', 'LotSiteAppeal', 'MailAddress', 'Main', 'Name', 'Oby', 'Pool', 
          'SaleData', 'TaxDistrict', 'TaxExemption', 'TypeConstruction', 'Value', 'VestingCodes']
```

```python
global_feature_list, global_entropy_list = ut.entropy_analysis_multiple_tables(tables)
```

```python
# plot the entropy distribution
import numpy as np
global_entropy_array = np.array(global_entropy_list)
global_feature_array = np.array(global_feature_list)
# remove 0 entropy terms
pos_global_entropy_array = global_entropy_array[global_entropy_array > 0]
pos_global_feature_array = global_feature_array[global_entropy_array > 0]
```

```python
pos_el = len(pos_global_entropy_array) # we eliminated a lot of zero terms!
```

```python
threshold = np.linspace(0, 14, 100)
count = []
for val in threshold:
    count.append(len(pos_global_entropy_array[pos_global_entropy_array >= val])/ pos_el)
frq_count = np.array(count)
```

```python
# plot frq_count vs. threshold
import matplotlib.pyplot as plt
# plotting the points
plt.plot(threshold, frq_count)
# naming the x axis
plt.xlabel('entropy threshold')
# naming the y axis
plt.ylabel('no. features with higher entropy than threshold')
# giving a title to my graph
plt.title('exceedence vs. threshold')
# function to show the plot
plt.show()
```

```python
# how to get the derivative of a discrete distributions
```

```python
threshold2 = np.linspace(0, 14, 20)
x_hist = []
hist = []
for i in range(1, len(threshold2)):
    low = threshold2[i - 1]
    high = threshold2[i]
    x_hist.append((low + high) / 2)
    result = np.where(np.logical_and(pos_global_entropy_array>= low, pos_global_entropy_array<= high))
    hist.append(len(result[0])/pos_el)
```

```python
# plot histogram of entropies
plt.plot(x_hist, hist)
# naming the x axis
plt.xlabel('entropy range')
# naming the y axis
plt.ylabel('no. features within entropy range')
# giving a title to my graph
plt.title('frequence vs. threshold')
# function to show the plot
plt.show()
```
